```python
#Librerias 
import time
import random
import sys
```

Definir variables

```python
#Escaleras 5x5| Casilla en la cual está la escalera:Casilla a la que sube
ladders= {}

ladders5x5 = {
    3:11,
    6:17,
    9:18,
    10:12
}

ladders6x6 = {
    4:12,
    7:18,
    10:19,
    11:13
}

#Serpientes 5x5| Casilla en la cual está la snake:Casilla a la cual baja
snakes = {}

snakes5x5 = {
    14:4,
    19:8,
    24:16,
    22:20   
}

snakes6x6 = {
    15:5,
    20:9,
    25:17,
    23:21   
}



Dado = 6 
ValorMax = 0
espera = 1

LadderText = [
    "subiendo la escalera",
    "encontraste una escalera, subiendo.."
]

SnakeText = [
    "serpiente encontrada",
    "para abajo como la serpiente que encontraste"
]

WinText = [
    "Ganaste la Champions",
    "Coronado como el campeón del jogo",
    "Pum pururum! con ustedes, el Campeón del juego, felicidades!"
]

TurnText = [
    "Tu turno.",
    "Hora de lanzar el dado.",
    "No hay tiempo extra como en futbol, lanza el dado."
]

```

```python
#BONUS: Mensaje de bienvenida para el juego.
def mensageinit():
    msg ='Bienvenido al juego escalera por Ethan para prueba test de Veevart'
    print(msg)

#Esta nos ayuda a definir un tamaño de tablero para iniciar el juego, 
#de acuerdo al tablero seleccionado, se organizan las snakes y ladders.
def tablero():
    print('Favor, seleccionar el tablero de preferencia (Escribir el numero que identifica al tablero)')
    eleccion = int(input('1-Tablero 5x5 | 2-Tablero 6x6'))
    if eleccion == 1:
        snakes = snakes5x5
        ladders = ladders5x5
        ValorMax = 25
        select = 'Seleccionaste tablero 5x5'
    elif eleccion == 2:
        snakes = snakes6x6
        ladders = ladders6x6
        ValorMax = 36
        select = 'Seleccionaste tablero 6x6'
    print(select)
    return snakes,ladders,ValorMax

#BONUS JUGADORES
def definir_jugadores():
    Numj = int(input("Indica la cantidad de jugadores"))
    jugadores = []
    for i in range(Numj):
        jugador = input("Favor, ingresar nombre de jugador")
        jugadores.append(jugador)
    
    print('Los jugadores de la partida son:')
    print(*jugadores, sep="\n")
    return jugadores
             

def get_valor_dado():
    time.sleep(espera)
    valor_dado = random.randint(1,Dado)
    print("Dado arroja" + ' ' + str(valor_dado))
    return valor_dado

def snake_ladder(jugador, current, dice_value):

    old_value = current

    #estamos definiendo si el valor viejo el cual es el antes de tirar los dados es + ó -, de esta manera,
    #indicaremos si los dados tienen que sumar o restar de acuerdo a la posición.
    if old_value > ValorMax:
        dice_value = -dice_value
        frase = "avanzar "
    elif old_value < ValorMax:
        dice_value = dice_value
        frase = "devolverte "
        
    print(dice_value)
    time.sleep(espera)
   
    current_value = current + dice_value
    comment = ''
    #esta se utiliza para saber cuanto le falta al usuario para ganar el juego BONUS
    if current_value > ValorMax or old_value > ValorMax:

        print("Necesitas "+ frase + str(abs(current_value - ValorMax)) + " para ganar el juego!")
        old_value = current_value
        return old_value
    
    
    #Avanza a la siguiente casilla
    print("\n" + jugador + " se mueve de " + str(old_value) + " a " + str(current_value))
    
    #Se identifica si cae en snake ó en ladder para aplicar la debida funcion
    if current_value in snakes:
        final_value = snakes.get(current_value)
        comment = jugador + " " + random.choice(SnakeText)

    elif current_value in ladders:
        final_value = ladders.get(current_value)
        comment = jugador + " " + random.choice(LadderText)

    else:
        #al no caer en ninguna, se sigue el juego de forma normal
        final_value = current_value
    print(comment)
    print("posición final de " + jugador + " es", final_value)
    

    return final_value

def check_win(jugador, position):
    time.sleep(espera)
    if ValorMax == position:
        print(random.choice(WinText))
        print("LO LOGRASTE!, " + jugador + ", ganaste el juego.")
        print('Gracias por jugar')
        sys.exit(1)
```

Logica juego

```python
def start():
    global snakes
    global ladders
    global ValorMax
    mensageinit()
    time.sleep(espera)
    snakes, ladders, ValorMax=tablero()
    time.sleep(espera)
    jugadores = definir_jugadores()
    time.sleep(espera)

    #Nos enfocamos en organizar las variables requeridas respecto a la posicion de los jugadores en el mapa
    
    position = []
    for j in range(len(jugadores)):
      pos = 0
      position.append(pos)
    
    #Inicio de juego
    while True:  
        time.sleep(espera)
        for k in range(len(jugadores)):
            jugador = jugadores[k]
            place = position[k]
            input_player = input('\n' + jugador + ': '+ random.choice(TurnText) + ' Dale ENTER para tirar los dados: ')
            print('\nLanzando Dados...')
            valor_dados = get_valor_dado()
            time.sleep(espera)
            print(jugador + ' moviendo espacios...')
            jugador_current_position = snake_ladder(jugador, place, valor_dados)
            #se guarda el valor actual de dicho jugador actualizando la list
            position[k] = jugador_current_position

            check_win(jugador, jugador_current_position)

 

 
```

```python
start()
```
